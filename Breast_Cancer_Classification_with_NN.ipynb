import dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
# loading the the data to a data frame
breast_cancer_dataset=sklearn.datasets.load_breast_cancer()
```

```python
print (breast_cancer_dataset)
```

```python
# load the data to a data frame
data_frame=pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
```

```python
# adding the taget column to the data frame
data_frame['label']=breast_cancer_dataset.target
```

seperating the features and target

```python
X=data_frame.drop(columns='label',axis=1)
Y=data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training and testing data

```python
X_train ,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_train_std=scaler.fit_transform(X_train)

X_test_std=scaler.transform(X_test)
```

**Buiding the neural network**

```python
# importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of neural network
model=keras.Sequential([
                        keras.layers.Flatten(input_shape=(30,)),
                        keras.layers.Dense(20,activation='relu'),
                        keras.layers.Dense(2,activation='sigmoid')

])
```

```python
# compiling the neural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']
)
```

```python
# training the neural network
history=model.fit(X_train_std,Y_train,validation_split=0.1,epochs=10)
```

Visulizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data','validation data'],loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data','validation data'],loc='upper right')
```

Accuracy of the model on test data

```python
loss ,accuracy=model.evaluate(X_test_std,Y_test)
print(accuracy)
```

Building the predictive system

```python
# Always provide input features in the same order as the training dataset.
# Changing the order can lead to wrong predictions, which is critical in medical ML.

input_data=(11.31,19.04,71.8,394.1,0.08139,0.04701,0.03709,0.0223,0.1516,0.05667,0.2727,0.9429,1.831,18.15,0.009282,0.009216,0.02063,0.008965,0.02183,0.002146,12.33,23.84,78,466.7,0.129,0.09148,0.1444,0.06961,0.24,0.06641
)

#change the input_data to a numpy array
input_data_as_numpy_array=np.asarray(input_data)

#reshape the numpy array as we are predicting for one data point
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)

#standardizing the input data
input_data_std=scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)

# Probability distribution
print("Prediction Probabilities:", prediction)

# Final class
prediction_label = np.argmax(prediction, axis=1)[0]
print("Predicted Class Label:", prediction_label)

# Meaning
if prediction_label == 0:
    print("🔴 The tumor is **Malignant**")
else:
    print("🟢 The tumor is **Benign**")
```
